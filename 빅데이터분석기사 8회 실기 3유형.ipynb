```python
#Q1. 로지스틱 회귀분석 

#작업형 3유형 1번
#survived를 종속변수, pclass, age, parch(부모 자식 총합), sibsp, fare 변수를 독립변수로 분석했을 때 아래 질문에 답하시오. (단, 반올림하여 소수점 둘째자리까지 구하시오.)

#1-1) 로지스틱 회귀분석 진행 중 유의하지 않은 변수의 수는? (독립변수는 총 15개이고 유의한 변수가 3개였다면 p-value값이 0.05보다 작은 유의하지 않은 변수는 12개)
#1-2) 유의한 변수로만 로지스틱 회귀분석을 진행한 후 회귀계수의 합계를 구하기 (단, 절편항도 포함한다.)
#1-3) 문제 2에서 구한 모델에서 A변수가 5단위 증가하면 오즈비는 몇 배로 변화하는가?

import pandas as pd
import numpy as np
import seaborn as sns

#Seaborn의 내장 타이타닉 데이터셋을 불러옵니다.
df=sns.load_dataset('titanic')
df=df[['survived','pclass','age','parch','sibsp','fare']]
df = df.dropna()

print(df.head())

#독립변수와 종속변수 설정 
x_train=df[['pclass','age','parch','sibsp','fare']]
y_train=df['survived']

#statsmodels 라이브러리 사용
#모델링(Logit함수)
import statsmodels.api as sm

x_train = sm.add_constant(x_train) #주의할점은 상수항을 추가해줘야 함 (8회 기출 지시사항: 상수항을 추가해라 <-> 추가하지 말아라라고 나오면 추가 X)
model_logit = sm.Logit(y_train,x_train).fit()#주의할 점은 y,x 순으로 입력해야 함.
summary = model_logit.summary()
print(summary)
#정답 1-1: 유의하지 안은 변수는 1개(fare P-value값이 0.194인 것)
```

```python

#유의한 변수만을 가지고 다시 모델링한다.
x_train=df[['pclass','age','parch','sibsp']]
y_train=df['survived']
import statsmodels.api as sm
x_train = sm.add_constant(x_train)
model_logit=sm.Logit(y_train, x_train).fit()
summary = model_logit.summary()
print(summary)

#회귀계수(coef) 출력한다.
print(model_logit.params)

#회귀계수(coef) 출력
print(round(sum(model_logit.params),2))
#문제1-2. 유의한 변수로만 로지스틱 회귀분석을 진행한 후 회귀계수의 합계를 구하기 정답: 2.45
```

```python
#문제 1-3. 만약 age변수가 5단위 증가하면 오즈비는 몇 배 증가하나요?
result=np.exp(5*model_logit.params)
print(round(result,2))
print(round(result[2],2))
#오즈비는 성공할 확률P를 성공하지 못할 확률 1-P로 나누는 것. exp^B0+B1x
#오즈변화량은 변화에서 기존으로 나누는 값 e^B0*e^B1(x+5)/ e^B0*e^B1x =따라서 e^5B1만큼 증가한다. 

#따라서 1-3의 정답은 오즈비가 0.8배 증가함을 알 수 있다.
```

```python
#Q2. 다중선형 회귀분석

#2-1)모델링 후 p-value가 가장 작은 변수의 회귀계수가 값을 구하시오.
#2-2) 1에서 구한 다중선형회귀 모델의 결정계수R을 구하시오.
#2-3) 1에서 만든 모델 A=( ), B=( ), C=( )의 값을 대입하면 Y값은 얼마로 예측되어야 하겠는가?

import pandas as pd
import numpy as np
import seaborn as sns

#Seaborn의 내장 타이타닉 데이터셋을 불러옵니다.
df=sns.load_dataset('titanic')
df=df[['survived','pclass','age','parch','sibsp','fare']]
df = df.dropna()

print(df.head())

#독립변수와 종속변수 설정 
x_train=df[['pclass','age','parch','sibsp','fare']]
y_train=df['survived']

import statsmodels.api as sm

#독립변수에서 상수항을 추가한다.
x=sm.add_constant(x_train)
y=y_train

#회귀모델 적합시킨다.
model = sm.OLS(y, x).fit()

print(model.summary())
#따라서 문제2-1)p-value가 가장 작은 변수의 회귀계수 값은 0.00값을 갖는 pclass의 -0.24이다. age의 -0.086

#문제 2-2. 1에서 구한 다중 선형 회귀 모델의 결정계수값은?
print("R-squared", model.rsquared) #정답: 0.19475

#문제 2-3. A=2, B=30, C=1, D=10, E=100을 대입할 때 Y의 값은? 
# 회귀계수 저장 (실제 값은 모델 summary 결과 참조)
const = -0.0301
coef_pclass = -0.1212
coef_age = -0.0051
coef_parch = 0.0850
coef_sibsp = -0.0330
coef_fare = 0.0025

# 예측값 계산
A = 2
B = 30
C = 1
D = 0
E = 100

y_pred = const + coef_pclass*A + coef_age*B + coef_parch*C + coef_sibsp*D + coef_fare*E
print("예측 생존 확률:", y_pred)
#정답: -0.09
```
