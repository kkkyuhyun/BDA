```python
#빅데이터분석기사 7회 실기 1유형
#문제1) 주어진 데이터 셋은 학생들의 과목별 점수 데이터이다.
#시험과목은 국어 영어 수학 컴퓨터이다.
#학생들은 이 중 3과목을 선택하여 들었으며 이에 대한 점수를 받았다
#가장 많은 학생이 들었던 과목을 구하고 결측치를 제거한 후 해당 과목의 점수를 표준화하여 가장 높은 점수를 소수 둘째자리까지 반올림하여 구하시오.

#데이터셋

import pandas as pd
import numpy as np

# 랜덤 시드 고정
np.random.seed(42)

# 1000개의 데이터 생성
num_students = 1000
data = {
    "student_id": np.arange(1, 1 + num_students),
    "국어": np.random.randint(60, 101, num_students),
    "영어": np.random.randint(60, 101, num_students),
    "수학": np.random.randint(60, 101, num_students),
    "컴퓨터": np.random.randint(60, 101, num_students),
}

students = pd.DataFrame(data)

# CSV 파일로 저장
students.to_csv("student_scores.csv", index=False)

students.head()

students.info()

#1. 가장 많은 학생이 들었던 과목 구하기
subject = students.iloc[:, 1:].count().idxmax()
subject #국어

#2. 해당 과목의 결측치를 제거하기
df = students.loc[:,subject].dropna()
df.info()
df.head()

#3. 표준화 Z-점수 공식: X-M/ 데이터 셋의 표준편차
#3-1)직접 계산
import numpy as np
df - df.mean()
score1 = (df - df.mean()) / df.std()
result1 = np.round(score1.max(),2)
result1

#3-2)사이킷런 활용
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df.values.reshape(-1,1))
score2 = scaler.transform(df.values.reshape(-1,1)) #2차원으로 변형
score2
result2 = np.round(score2.max(),2)
result2

#3-3) stats코드 
from scipy.stats import zscore
import numpy as np
score3 = zscore(df)
score3
result3 = np.round(score3.max(),2)
result3
```
